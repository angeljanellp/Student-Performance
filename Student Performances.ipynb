## Data Preparation

```python
# no need to encode as they are a numerical columns
df.dtypes
```

```python
# no need to fill NaN since there are no missing values
df.isnull().sum()
```

```python
#due to the data being in different ranges, we will scale the data using Standard Scalers
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(df)
```

### Model 1: Kmeans Clustering

```python
#Create multiple models and do an elbow graph to find the best n_cluster value

from sklearn.cluster import KMeans
ssd=[]
for k in range(2,50):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    ssd.append(kmeans.inertia_)
```

```python
plt.plot(range(2,50), ssd, marker='o')
plt.title('best n_cluster Value');
plt.xlabel('N_cluster VALUE')
plt.ylabel('inertia')
plt.grid(True)
```

```python
#it looks like here the elbow is between 10-15
```

```python
#creating a model with a n_cluster of 12
kmeans = KMeans(n_clusters=12, random_state=56)
kmeans.fit(X_scaled)
```

### Model 2: DBSCAN 

```python
#first create multiple models to find the best eps value, while considering the outliers
from sklearn.cluster import DBSCAN
outliers=[]
eps=np.linspace(0.001,5,50)
for ep in eps:
    dbscan=DBSCAN(eps=ep,min_samples=3)
    dbscan.fit(X_scaled) 
    
    labels = dbscan.labels_
    outlier = (len(df[labels == -1])/len(df))*100
    outliers.append(outlier)
    
```

```python
plt.plot(np.linspace(0.001,5,50), outliers, marker="o", )
plt.title('Best Eps Value')
plt.xlabel('EPS VALUE')
plt.ylabel('percent')
plt.grid(True)
```

```python
# looking at the values better with a series
pd.Series(data=outliers,index=eps)
```

```python
# here we see that the elbow is situated between 2 and 3
# we will be choosing the eps value of 2.70

dbscan=DBSCAN(eps=2.70,min_samples=2)
dbscan.fit(X_scaled)
```

## Clustering with both models

```python
#adding cluster labels
df['Kmeans']=kmeans.labels_
df['DBSCAN']=dbscan.labels_
```
